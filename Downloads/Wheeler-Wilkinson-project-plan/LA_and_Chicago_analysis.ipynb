# Cleaning

The Chicago dataset "Date" refers to the date and time that the event occured. Therefore we will also use the date and time the event occurd (DATE OCC) from the LA Dataset. Using datetime we formated the DATE OCC to match that of Chicago and droped all row that were not from the year 2020.

```python
LA['DATE OCC'] = pd.to_datetime(LA['DATE OCC'])

LA = LA[LA['DATE OCC'].dt.year == 2020]

# Reformat to match the Chicago format: %m/%d/%Y %I:%M:%S %p
LA['DATE OCC'] = LA['DATE OCC'].dt.strftime('%m/%d/%Y %I:%M:%S %p')

LA.head(5)
```

# Save Cleaned DataFrames Back to CSV
After cleaning the dataframes, we save them as cleaned CSV files to upload to GitHub since they will be much smaller after cleaning.

```python
# Save Cleaned DataFrames to New CSV Files
# This saves the cleaned data as new files without overwriting the originals.

# Save Chicago dataframe to a new file
chicago.to_csv('Chicago_Crime_Cleaned.csv', index=False)
print("Chicago cleaned CSV saved as 'Chicago_Crime_Cleaned.csv'.")

# Save LA dataframe to a new file
LA.to_csv('LA_Crime_Cleaned.csv', index=False)
print("LA cleaned CSV saved as 'LA_Crime_Cleaned.csv'.")
```
