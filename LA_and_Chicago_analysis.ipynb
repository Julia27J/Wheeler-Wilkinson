# Cleaning

The Chicago dataset "Date" refers to the date and time that the event occured. Therefore we will also use the date and time the event occurd (DATE OCC) from the LA Dataset. Using datetime we formated the DATE OCC to match that of Chicago and droped all row that were not from the year 2020.

```python
LA['DATE OCC'] = pd.to_datetime(LA['DATE OCC'])

LA = LA[LA['DATE OCC'].dt.year == 2020]

# Reformat to match the Chicago format: %m/%d/%Y %I:%M:%S %p
LA['DATE OCC'] = LA['DATE OCC'].dt.strftime('%m/%d/%Y %I:%M:%S %p')

LA.head(5)
```

Gemini generated Python code using a dictionary inversion method to recategorize both datasets. Convert the Data column from string to datetime objects and sort the data frame by date so that the rows are all listed in order of occurrence. We also reset the index so it's clean (0, 1, 2...) after sorting.

```python
import pandas as pd

# 1. Convert the 'Date' column from strings to datetime objects
combined_crime['Date'] = pd.to_datetime(combined_crime['Date'])

# 2. Sort the DataFrame by 'Date' (ascending=True means oldest to newest)
combined_crime = combined_crime.sort_values(by='Date', ascending=True)

# 3. Optional: Reset the index so it's clean (0, 1, 2...) after sorting
combined_crime = combined_crime.reset_index(drop=True)

combined_crime['Date'] = combined_crime['Date'].dt.normalize()

# Check the results
combined_crime.head(5)
```

**Data Dictionary**

**Chicago Dataset**

**`ID`**

- `[ID]` a unique number associated with each crime as an identifier

**`Date`**

- `[Date]` the date the crime occurred, in month/day/year format, including the time

**`Latitude`**

- `[Latitude]` the latitude coordinate of the location where the crime occurred 
**`Longitude`**

- `[Longitude]` the longitude coordinate of the location where the crime occurred 

**`Crime Type`**

- `[Unified_Crime_Type]` the classification of the type of crime that occurred

**LA Dataset**

**`Division of Records Number`**

- `[DR_NO]` an identification number for each crime

**`Date Occurred`**

- `[DATE OCC]` the date in which the crime occurred in month/date/year format, including the time

**`Latitude`**

- `[LAT]` the latitude coordinate of the location where the crime occurred 

**`Longitude`**

- `[LON]` the longitude coordinate of the location where the crime occurred
